```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np
#change from float64 to int64
scaler = MinMaxScaler()

CPO_df_sub.head(), CPO_df_sub.tail()
```

```python
scaler = MinMaxScaler()
data_train = scaler.fit_transform(train.reshape(-1, 1))
data_test = scaler.transform(test.reshape(-1, 1))
```

```python
def train_preparation(data):
    x = []
    y = []
    
    for i in range(1, len(data)):
        x.append(data[i-1:i, 0])
        y.append(data[i])
        
    x = np.array(x)
    y = np.array(y)
    
    x = x.reshape(x.shape[0], x.shape[1], 1)
    
    return x, y
```

```python
x_train, y_train = train_preparation(data_train)
```

```python
def test_preparation(data):
    x = []
    y = []
    
    for i in range(1, len(data)):
        x.append(data[i-1:i, 0])
        y.append(data[i])
        
    x = np.array(x)
    y = np.array(y)
    
    x = x.reshape(x.shape[0], x.shape[1], 1)
    
    return x, y
```

```python
x_test, y_test = test_preparation(data_test)
```

```python
from keras.models import Sequential
from tensorflow.keras.layers import Bidirectional
from keras.layers import LSTM,Dense ,Dropout
```

```python
model = Sequential()
model.add(Bidirectional(LSTM(units=250, return_sequences=True, input_shape = (x_train.shape[1],1))))
model.add(Dropout(0.2))
model.add(Bidirectional(LSTM(units= 250 , return_sequences=True)))
model.add(Dropout(0.2))
model.add(Bidirectional(LSTM(units= 250 , return_sequences=True)))
model.add(Dropout(0.2))
model.add(Bidirectional(LSTM(units= 250 , return_sequences=True)))
model.add(Dropout(0.2))
model.add(Bidirectional(LSTM(units= 250 , return_sequences=True)))
model.add(Dropout(0.2))
model.add(Bidirectional(LSTM(units= 250)))
model.add(Dropout(0.2))
model.add(Dense(1,activation='linear'))
```

```python
# regressor = Sequential()
# regressor.add(Bidirectional(LSTM(units=30, return_sequences=True, input_shape = (x_train.shape[1],1) ) ))
# regressor.add(Dropout(0.2))
# regressor.add(LSTM(units= 30 , return_sequences=True))
# regressor.add(Dropout(0.2))
# regressor.add(LSTM(units= 30 , return_sequences=True))
# regressor.add(Dropout(0.2))
# regressor.add(LSTM(units= 30))
# regressor.add(Dropout(0.2))
# regressor.add(Dense(units = 1,activation='linear'))
# #I have used Adam optimizer because it is computationally efficient.
model.compile(optimizer='adam', loss='mean_squared_error',metrics=['acc'])
model.fit(x_train, y_train, epochs=30,batch_size=32 )
```

```python
y_pred = model.predict(x_test)
```
